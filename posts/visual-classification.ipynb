# Visual Classification of Engine Diagnostic Events with Convoluted Neural Networks



```python
%matplotlib inline
from control.matlab import *
import matplotlib.pyplot as plt
```

```python
s = tf([1, 0],[1])
```

```python
K   = 1
tau = 1
sys1= K/(tau*s+1)
```

```python
yout, T = step(sys1)
plt.plot(T, yout)
```
